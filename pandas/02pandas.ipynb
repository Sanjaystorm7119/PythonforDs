```python
#new columns using assign()

#creating values for columns
valuesCol1 = np.random.randint(1,101, size=len(airport_df))
valuesCol2 = np.random.randint(1,101, size=len(airport_df)) 
```

```python
#creating series using random value col
series1 = pd.Series(valuesCol1,index=np.arange(0,len(airport_df)))
series2 = pd.Series(valuesCol2,index=np.arange(0,len(airport_df)))

airport_df.assign(valuesCol1=series1, valuesCol2=series2)
# airport_df - the assign only copies so its values in not updates in airport_df
airport_df = airport_df.assign(valuesCol1=series1, valuesCol2=series2) #updating airport_df
airport_df

```
